```python
# Getting the data in the session
from sklearn.datasets import load_iris #scikit-learn dataset

# Load the dataset
iris = load_iris()

# Access features and target labels
X = iris.data  # Feature matrix
y = iris.target  # Target classes
```

```python
# Create DataFrame from features (X)
df = pd.DataFrame(X, columns=iris.feature_names) # pd is an alias for pandas, DataFrame is a 2D labeled data structure with columns of potentially different types. It is similar to a spreadsheet or SQL table, or a dict of Series objects.

# Add the target (y) as a new column
df['target'] = y

print(df.head())
```
